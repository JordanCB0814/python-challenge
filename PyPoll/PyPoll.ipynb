```python
#Our Dependencies
import csv
import os


#Desired Outcome
#Election Results
#-------------------------
#Total Votes: 369711
#-------------------------
#Charles Casper Stockham: 23.049% (85213)
#Diana DeGette: 73.812% (272892)
#Raymon Anthony Doane: 3.139% (11606)
#-------------------------
#Winner: Diana DeGette
#-------------------------

#Files that we load/send our data to
file_to_load = os.path.join(".","Resources","election_data.csv")
file_to_output = os.path.join(".","election_analysis.txt")

#Variables being evaluated and reset
total_votes = 0
candidate_votes = {}
candidate_options = []
winning_candidate = ""
winning_counter = 0

with open(file_to_load) as election_data:
        reader = csv.reader(election_data)
        header = next(reader)
    
        for row in reader:
            total_votes = total_votes + 1

            candidate_names = row[2]

            if candidate_names not in candidate_options:
                candidate_options.append(candidate_names)

                candidate_votes[candidate_names]= 0
            candidate_votes[candidate_names] += 1


with open(file_to_output, "w")as txt_file:
    poll_results = (
            f"Election Results\n"
            f"-------------------------\n"
            f"Total Votes:{total_votes}\n"
            f"-------------------------\n"     
    )
    
    print(poll_results, end="")
    txt_file.write(poll_results)

    for candidate in candidate_votes:
        votes = candidate_votes[candidate]
        votes_percentage = float(votes) / float (total_votes) * 100
       
        if (votes > winning_counter):
            winning_counter = votes
            winning_candidate = candidate
        voter_results = f"{candidate}: {votes_percentage:.3f}% ({votes})\n"
        print(voter_results, end="")

        txt_file.write(voter_results)

    winning_candidate_summary = (
        f"-------------------------\n"
        f"Winner:{winning_candidate}\n"
        f"-------------------------\n"

)
    print(winning_candidate_summary)

    txt_file.write(winning_candidate_summary)
        
```
